```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import torch
from inverse_gamma import InverseGamma
import numpy as np
```

```python
alpha = torch.tensor(3.)
beta = torch.tensor(0.5)
p = InverseGamma(alpha, scale=beta)
alpha = torch.tensor(3.)
beta = torch.tensor(1.)
p2 = InverseGamma(alpha, scale=beta)
alpha = torch.tensor(2.)
beta = torch.tensor(1.)
p3 = InverseGamma(alpha, scale=beta)
```

```python
result1 = []
result2 = []
result3 = []
for i in range(1000):
    result1.append(p.rsample().item())
    result2.append(p2.rsample().item())
    result3.append(p3.rsample().item())
```

```python
plt.xlim(0, 3)
sns.distplot(np.array(result1), label=r"$\alpha=3, \beta=0.5$")
sns.distplot(np.array(result2), label=r"$\alpha=3, \beta=1$")
sns.distplot(np.array(result3), label=r"$\alpha=2, \beta=1$")
plt.legend()
```
